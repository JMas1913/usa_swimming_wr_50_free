* We can see that index 11 and 12 are duplicates and therefore, we should remove the duplicate
* We also see that the swimmers who's club code is USA do not have a country. We should remove as many `NaN` values as possible because missing data is not a good thing. It is safe to assume that the swimmers with a club code as USA have a country as United States

#### Removing the duplicates

```python
# Remove duplicates
men_wr_50_free = men_wr_50_free.drop_duplicates()
# Reset the index
men_wr_50_free = men_wr_50_free.reset_index(drop=True)
# Save the changes
men_wr_50_free.to_csv('/Users/mastinfitness/Desktop/Documents/Data_Analytics/USA_Swimming/Men_World_Record_Progression_50_Free.csv')
```

```python
# Verifying the duplicate was dropped properly
men_wr_50_free
```

The duplicate was in fact dropped properly.
* Now to fix the `NaN` values.

```python
# Find the rows where the Club Code is USA and the Swimmer Country is NaN
rows_to_update = men_wr_50_free[(men_wr_50_free['CLUB CODE'] == 'USA') & (men_wr_50_free['SWIMMER\nCOUNTRY'].isnull())]
# Fill in the NaN values with United States
rows_to_update['SWIMMER\nCOUNTRY'] = 'United States'
# Update the data
men_wr_50_free.update(rows_to_update)
# Save the data
men_wr_50_free.to_csv('/Users/mastinfitness/Desktop/Documents/Data_Analytics/USA_Swimming/Men_World_Record_Progression_50_Free.csv')
```

```python
# Validating that the `NaN` values have been changed to United States where the CLUB CODE = USA
men_wr_50_free
```

We repeat the previous step with the UNK `CLUB CODE` and fill in the `NaN` with `Unknown`.

```python
# Find the rows where the Club Code is UNK and the Swimmer Country is NaN
rows_to_update = men_wr_50_free[(men_wr_50_free['CLUB CODE'] == 'UNK') & (men_wr_50_free['SWIMMER\nCOUNTRY'].isnull())]
# Fill in the NaN values with Unknown
rows_to_update['SWIMMER\nCOUNTRY'] = 'Unknown'
# Update the data
men_wr_50_free.update(rows_to_update)
# Save the data
men_wr_50_free.to_csv('/Users/mastinfitness/Desktop/Documents/Data_Analytics/USA_Swimming/Men_World_Record_Progression_50_Free.csv')
```

Now that the issues have been fixed we can move on to answering a few questions.
* Which country has the most world records for the 50 Free for Men and Women?
* Which athlete has the most world records for the 50 Free for Men and Women?

Without more data we cannot ask more questions because there are too many variables.

**Notes:** There are more rows for the Men's 50 Free than there is for the Women's 50 Free.

#### Let's work on the Men's 50 Free Data

```python
# Men's 50 Free: Which country has the most world records?

# Initialize an empty dictionary to store the country's name and their count
country_counts = {}

# Iterate through each row in the DataFrame
for index, country in men_wr_50_free.iterrows():
    # Extract the country's name
    athlete_country = country['SWIMMER\nCOUNTRY']

    # Check if the country's name already exists in the dictionary
    if athlete_country in country_counts:
        # If it does, increment their count by 1
        country_counts[athlete_country] += 1
    else:
        # If it doesn't, add their name to the dictionary with a count of 1
        country_counts[athlete_country] = 1

# Iterate through each country's name and their count in the dictionary
print('Which Country Has The Most World Records?')
for athlete_country, count in country_counts.items():
    # Display the athlete's name and their count
    print(f"{athlete_country}: {count}")
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='SWIMMER\nCOUNTRY', data=men_wr_50_free)
plt.xlabel('SWIMMER\nCOUNTRY')
plt.ylabel('Count')
plt.title('Count of Each Country in the Dataset')
plt.show()
```

```python
# Men's 50 Free: Which Athlete has the most world records?

# Create a dictionary to store the athlete names and their counts
athlete_counts = {}

# Iterate through the rows of the DataFrame
for index, row in men_wr_50_free.iterrows():
    # Retrieve the athlete name and update the corresponding count in the dictionary
    athlete_name = row['ATHLETE']
    if athlete_name in athlete_counts:
        athlete_counts[athlete_name] += 1
    else:
        athlete_counts[athlete_name] = 1

# Display the athlete's name along with their count
print("Which Athlete Has The Most World Records?")
for athlete_name, count in athlete_counts.items():
    print(f"{athlete_name}: {count}")
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='ATHLETE', data=men_wr_50_free)
plt.xlabel('ATHLETE')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.title('Count of Each Athlete in the Dataset')
plt.show()
```
